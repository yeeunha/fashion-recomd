```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_1 = pd.read_csv('/content/drive/MyDrive/데청캠_공유/쇼핑몰 라벨/러블리_revised.csv')
```

```python
# 쇼핑몰 별 가장 높은 빈도의 'style' 종류와 비율 찾기
def most_common_with_ratio(series):
    counts = series.value_counts()
    most_common_style = counts.idxmax()
    most_common_ratio = counts.max() / counts.sum()
    return pd.Series({'most_common_style': most_common_style, 'most_common_ratio': most_common_ratio})

most_common_info = df_1.groupby('mall')['style'].apply(most_common_with_ratio)

print(most_common_info)
```
